### First, Set up environment

```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

import caffe

# Set caffe to use GPU
caffe.set_device(0)
caffe.set_mode_gpu()
```

Set up the data from the csv files, train.csv and test.csv
Images are 28x28 grayscale
train.csv has 785 columns, first column is digit label, following columns are pixel values from 0 to 255
test.csv has 784 columns, same as train.csv but without the label column

```python
total_train_set = pd.read_csv("DigitData\\train.csv")

train_data = (total_train_set.ix[:,1:].values).astype('float32')
train_labels = total_train_set.ix[:,0].values.astype('float32')
```

```python
train_data = train_data.reshape(train_data.shape[0],  28, 28)
train_data = train_data[:,np.newaxis,:,:]
print "Data Shape: %s" % (str(train_data.shape))
print "Label Shape: %s" % (str(train_labels.shape))

print "Sample Images:"
for i in range(0,5):
    plt.subplot(1,5,i+1)
    plt.imshow(train_data[i,0], cmap=plt.get_cmap('gray'))
plt.show()
```

Then split the train data into train(70%) and cross validation(30%) sets

```python
cross_validation_percent = 20.0

cv_data = train_data[0:int(cross_validation_percent/100 * len(train_data)), :, :]
cv_labels = train_labels[0:int(cross_validation_percent/100 * len(train_labels))]

train_data = train_data[int(cross_validation_percent/100 * len(train_data)):, :, :]
train_labels = train_labels[int(cross_validation_percent/100 * len(train_labels)):]
```

## Define the network, LeNet architecture

```python
from caffe import layers as L, params as P

def lenet(batch_size, train=True, labels=True):
    n = caffe.NetSpec()
    
    
    n.data, n.label = L.MemoryData(batch_size=batch_size, channels=1,transform_param=dict(scale=1./255), height=28, width=28, ntop=2)
    n.conv1 = L.Convolution(n.data, kernel_size=5, num_output=20, weight_filler=dict(type='xavier'))
    n.pool1 = L.Pooling(n.conv1, kernel_size=2, stride=2, pool=P.Pooling.MAX)
    n.conv2 = L.Convolution(n.pool1, kernel_size=5, num_output=50, weight_filler=dict(type='xavier'))
    n.pool2 = L.Pooling(n.conv2, kernel_size=2, stride=2, pool=P.Pooling.MAX)
    n.fc1 =   L.InnerProduct(n.pool2, num_output=500, weight_filler=dict(type='xavier'))
    n.relu1 = L.ReLU(n.fc1, in_place=True)
    n.score = L.InnerProduct(n.relu1, num_output=10, weight_filler=dict(type='xavier'))
    
    if labels:
        n.acc = L.Accuracy(n.score, n.label)
        n.loss =  L.SoftmaxWithLoss(n.score, n.label)
    else:
        n.prob = L.Softmax(n.score)
    
    return n.to_proto()
```

## Define solver (tweaked from fine-tune example)

```python
from caffe.proto import caffe_pb2

def solver(train_net_path, test_net_path=None, base_lr=0.01):
    s = caffe_pb2.SolverParameter()

    # Specify locations of the train and (maybe) test networks.
    s.train_net = train_net_path
    if test_net_path is not None:
        s.test_net.append(test_net_path)
        s.test_interval = 1000  # Test after every 1000 training iterations.
        s.test_iter.append(100) # Test on 100 batches each time we test.

    # The number of iterations over which to average the gradient.
    # Effectively boosts the training batch size by the given factor, without
    # affecting memory utilization.
    #s.iter_size = 1
    
    s.max_iter = 100000     # # of times to update the net (training iterations)
    
    # Solve using the stochastic gradient descent (SGD) algorithm.
    # Other choices include 'Adam' and 'RMSProp'.
    s.type = 'Adam'

    # Set the initial learning rate for SGD.
    s.base_lr = base_lr

    s.lr_policy = 'step'
    s.gamma = 0.5
    #s.power=0.75
    s.stepsize = 400

    # Set other SGD hyperparameters. Setting a non-zero `momentum` takes a
    # weighted average of the current gradient and previous gradients to make
    # learning more stable. L2 weight decay regularizes learning, to help prevent
    # the model from overfitting.
    s.momentum = 0.9
    s.weight_decay = 5e-3

    # Display the current training loss and accuracy every 1000 iterations.
    s.display = 1000

    # Snapshots are files used to store networks we've trained.  Here, we'll
    # snapshot every 10K iterations -- ten times during training.
    s.snapshot = 10000
    s.snapshot_prefix = 'test'
    
    # Train on the GPU.  Using the CPU to train large networks is very slow.
    s.solver_mode = caffe_pb2.SolverParameter.GPU
    
    # Write the solver to a temporary file and return its filename.
    with open('solver.prototxt', 'w') as f:
        f.write(str(s))
        return f.name
```

```python
def run_solvers(niter, solvers, batch_size, disp_interval=10):
    """Run solvers for niter iterations,
       returning the loss and accuracy recorded each iteration.
       `solvers` is a list of (name, solver) tuples."""
    blobs = ('loss', 'acc')
    loss, acc = ({name: np.zeros(niter) for name, _ in solvers}
                 for _ in blobs)
    for it in range(niter):
        solver.net.set_input_arrays(np.ascontiguousarray(train_data[it*batch_size:(it+1)*batch_size,:,:,:]), np.ascontiguousarray(train_labels[it*batch_size:(it+1)*batch_size]))  
        for name, s in solvers:
            s.step(1)  # run a single SGD step in Caffe
            loss[name][it] = s.net.blobs['loss'].data.copy()
            acc[name][it] = s.net.blobs['acc'].data.copy()
        if it % disp_interval == 0 or it + 1 == niter:
            loss_disp = '; '.join('%s: loss=%.3f, acc=%2d%%' %
                                  (n, loss[n][it], np.round(100*acc[n][it]))
                                  for n, _ in solvers)
            print '%3d) %s' % (it, loss_disp)
    return loss, acc
```

Write the nets to file and set batch sizes and the number of iterations to train and test.



```python
# TRAIN NET
#-------------------------------------------------
batch_size = 64
epochs = 4
niter = int(len(train_data) / batch_size)*epochs  # number of iterations to train

# write the train net to file
with open('lenet.prototxt', 'w') as f:
    f.write(str(lenet(batch_size, train=True)))

# TEST NET
#-------------------------------------------------
test_batch_size = 64
test_niter = int(len(cv_data) / test_batch_size)

# Ensure that do not go outisde cross validation set bounds
assert test_niter*test_batch_size <= cv_data.shape[0]

with open('lenet_test.prototxt', 'w') as f:
    f.write(str(lenet(test_batch_size, train=False)))
```

Now run the solver for 'niter' iterations

```python
solver_filename = solver('lenet.prototxt', test_net_path = 'lenet_test.prototxt', base_lr=0.001)
solver = caffe.get_solver(solver_filename)

test_interval = 10 # how many iterations in between tests
disp_interval = 10 # how many iterations between display

print 'Running solvers for %d iterations...' % (niter)
solvers = [('lenet', solver)]
blobs = ('loss', 'acc')
loss, acc = ({name: np.zeros(niter) for name, _ in solvers} for _ in blobs)

num_test_iterations = niter/test_interval + 1 if (niter % test_interval == 0) else (niter/test_interval + 1)
             
test_loss, test_acc = ({name: np.zeros(num_test_iterations) for name, _ in solvers}
             for _ in blobs)
solver.test_nets[0].set_input_arrays(np.ascontiguousarray(cv_data[0*test_batch_size:(0+1)*test_batch_size,:,:,:]), np.ascontiguousarray(cv_labels[0*test_batch_size:(0+1)*test_batch_size]))

test_it = 0

for it in range(niter):
    solver.net.set_input_arrays(np.ascontiguousarray(train_data[(it%459)*batch_size:((it%459)+1)*batch_size,:,:,:]), np.ascontiguousarray(train_labels[(it%459)*batch_size:((it%459)+1)*batch_size]))  
    
    for name, s in solvers:
        s.step(1)  # run a single SGD step in Caffe
        loss[name][it] = s.net.blobs['loss'].data.copy()
        acc[name][it] = s.net.blobs['acc'].data.copy()
        
        if it % test_interval == 0 or it + 1 == niter:
            for it_t in range(test_niter):
                s.test_nets[0].set_input_arrays(np.ascontiguousarray(cv_data[it_t*test_batch_size:(it_t+1)*test_batch_size,:,:,:]), np.ascontiguousarray(cv_labels[it_t*test_batch_size:(it_t+1)*test_batch_size]))
                #net.set_input_arrays(np.ascontiguousarray(cv_data[0:test_niter*test_batch_size,:,:,:]), np.ascontiguousarray(cv_labels[0:test_niter*test_batch_size]))
                s.test_nets[0].forward()
            
            test_loss[name][test_it] = s.test_nets[0].blobs['loss'].data.copy()
            test_acc[name][test_it] = s.test_nets[0].blobs['acc'].data.copy()
            test_it += 1
            
    if it % disp_interval == 0 or it + 1 == niter:
        loss_disp = '; '.join('%s: loss=%.3f, acc=%2d%%' % (n, loss[n][it], np.round(100*acc[n][it])) for n, _ in solvers)
        print '%d) %s' % (it, loss_disp)

print 'Done.'

train_loss = loss['lenet']
train_acc = acc['lenet']
test_loss = test_loss['lenet']
test_acc = test_acc['lenet']
```

## Visualize First Conv Layer Filters

Here we show a visualization of the first convolutional layer filters, using vis_square function from the caffe classification example.

```python
def vis_square(data):
    """Take an array of shape (n, height, width) or (n, height, width, 3)
       and visualize each (height, width) thing in a grid of size approx. sqrt(n) by sqrt(n)"""
    
    # normalize data for display
    data = (data - data.min()) / (data.max() - data.min())
    
    # force the number of filters to be square
    n = int(np.ceil(np.sqrt(data.shape[0])))
    padding = (((0, n ** 2 - data.shape[0]),
               (0, 1), (0, 1))                 # add some space between filters
               + ((0, 0),) * (data.ndim - 3))  # don't pad the last dimension (if there is one)
    data = np.pad(data, padding, mode='constant', constant_values=1)  # pad with ones (white)
    
    # tile the filters into an image
    data = data.reshape((n, n) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((n * data.shape[1], n * data.shape[3]) + data.shape[4:])
    
    plt.imshow(data, cmap='gray'); plt.axis('off')
    

net = caffe.Net('lenet.prototxt', 'weights.caffemodel', caffe.TEST)
    
# the parameters are a list of [weights, biases]
filters = net.params['conv1'][0].data
vis_square(np.squeeze(filters.transpose(0, 2, 3, 1)))
plt.show()
```

## Visualize what images the network misclassified

Run the net on the cross validation set.  See which numbers are most often misclassified, and what are they misclassified as.

```python
# set up array to store log of label to classifcation result
# 10x10 array, rows correspond to correct label, columns correspond to classifcation result
# each array location contains the number of times each label was classified as that result
classification_results = np.zeros((10,10))

for it in range(test_niter):
    net.set_input_arrays(np.ascontiguousarray(cv_data[it*test_batch_size:(it+1)*test_batch_size,:,:,:]), np.ascontiguousarray(cv_labels[it*test_batch_size:(it+1)*test_batch_size]))
    net.forward()
    # Record classification
    classifications = np.argmax(net.blobs['score'].data, 1)
    for i, c in enumerate(classifications):
        classification_results[int(cv_labels[it*test_batch_size+i])][c] += 1 # tally the classification result into the proper label row
```

```python
# Plot the results
percent_correct = [(classification_results[i][i] / x) * 100 for i,x in enumerate(sum(classification_results,0))]

fig, ax = plt.subplots()
ax.bar(np.arange(len(classification_results)), percent_correct)
ax.set_xticks(np.arange(len(classification_results)))
ax.set_xlabel("Number")
ax.set_ylabel("Accuracy Percent")
ax.set_title("Classification Accuracy")
plt.show()
```

## Now let's see what they were misclassified as!

Draw pie charts for each digit (0,1,2 ..) showing which digits they were misclassified as

```python
matplotlib.rcParams['figure.figsize'] = [15,7]
fig, ax = plt.subplots(nrows=2,ncols=5)
colors = ['yellowgreen','red','lightskyblue','white','lightcoral','blue','pink', 'darkgreen','yellow','violet']
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.title(str(i))
    labels = [r for r in range(10) if r != i]
    patches, texts = plt.pie(classification_results[i][labels], colors=[colors[i] for i in labels], radius=3)
    for t in texts:
        t.set_size('larger')
    plt.axis('equal')
    
for i, c in enumerate(colors):
    print "%d: %s " % (i,c)

plt.tight_layout()
plt.show()
```

## Output the labels to kaggle format

Run the test data from Kaggle through the trained network and put the output labels into a csv file that Kaggle can read.

```python
import pandas as pd
import numpy as np
import caffe
# 28000 exampes in test_set
test_set = pd.read_csv("DigitData\\test.csv")
test_set = test_set.as_matrix().reshape(test_set.shape[0],  28, 28)
test_set = test_set[:,np.newaxis,:,:]
test_set = np.ascontiguousarray(test_set)

test_batch_size = 64

# Set up the net
with open('lenet_out.prototxt', 'w') as f:
    f.write(str(lenet(test_batch_size, train=False, labels=False)))

net_out = caffe.Net('lenet_out.prototxt', 'weights.caffemodel', caffe.TEST)

```

```python
# Open the output file
f = open('DigitData\\out.csv', 'w')
f.write('ImageId,Label\n')

labels = np.zeros(test_batch_size) # stores the classified labels
zeros = np.zeros(test_batch_size).astype('float32') # used as dummy labels, network wouldn't initialize without having two tops on memory data layer
iters = len(test_set) / test_batch_size

for i in range(iters):
    net_out.set_input_arrays((test_set[i*test_batch_size:(i+1)*test_batch_size]).astype('float32'), zeros)
    net_out.forward()
    labels = np.argmax(net_out.blobs['prob'].data, 1)
    for it in range(len(labels)):
        f.write('%d,%d\n' % ((i*test_batch_size + it) + 1, labels[it]))
        

# Train on the leftovers
leftovers = len(test_set) % test_batch_size
padded_data = np.zeros((test_batch_size, test_set.shape[1], test_set.shape[2], test_set.shape[3]))
padded_data[0:leftovers] = test_set[-leftovers:]

if leftovers != 0:
    net_out.set_input_arrays(padded_data.astype('float32'), np.zeros(test_batch_size).astype('float32'))
    net_out.forward()
    labels = np.argmax(net_out.blobs['prob'].data, 1)
    for it in range(leftovers):
        f.write('%d,%d\n' % (iters*test_batch_size + it + 1, labels[it]))
    
f.close()
```
